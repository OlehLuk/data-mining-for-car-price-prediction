```python
import pandas as pd
df = pd.read_csv('all_null_replaced_or_deleted.csv', encoding='utf-8')
```

```python
df.sample(5)
```

```python
test_df = pd.DataFrame(df, copy=True)
```

```python
rare_bs = test_df.groupby('bodyStyle').count()
rare_bs.loc[(rare_bs['id'] < 2000) ]
```

### Assign class other to all rare bodyStyles

```python
bodyStyle_temp = test_df.groupby('bodyStyle').count() 
```

```python
bodyStyle_temp = test_df.groupby('bodyStyle').count()['id']
body_style_to_replace = bodyStyle_temp[bodyStyle_temp < 5].index
```

```python
test_df.loc[test_df['bodyStyle'] == 'drugoj', 'bodyStyle'] = 'other'
```

```python
for bs in body_style_to_replace:
    test_df.loc[test_df['bodyStyle'] == bs, 'bodyStyle'] = 'other'
```

```python
test_df.groupby('bodyStyle').count()[['id']]
```

```python
test_df.groupby(['drive']).count()[['id']]
```

## Get rid of rare marks
### Statistics for them

```python
rare_marks = test_df.groupby('mark').count()[test_df.groupby('mark').count()['id'] <= 5]['id']
print(str(rare_marks.shape[0]) + ' marks have <= 5 cars, are rare')
```

```python
rare_marks_names = rare_marks.index.get_values()
print(rare_marks_names)
```

### Replace with 'other'

```python
for bs in rare_marks_names:
    test_df.loc[test_df['mark'] == bs, ['mark','model']] = ['other', 'other']
    
```

```python
other_df = test_df[test_df['mark'] == 'other']
other_df
```

```python
other_df.loc[:,['price', 'race', 'year','engineVolume']].describe()
```

## Getting rid of rare models

```python
rare_models = test_df.groupby(['mark', 'model']).count()[test_df.groupby(['mark', 'model']).count()['id'] <= 5]['id']
print(str(rare_models.count()) + ' models have <= 5 cars, are rare')
```

```python
rare_models
```

### Get statistics for this rare cars

```python
rare_models.shape[0]
```

```python
rare_models.groupby('mark').count()
```

### Assign other to rare marks / models

```python
rare_models_list = rare_models.index.tolist()
```

```python
for bs in rare_models_list:
    test_df.loc[(test_df['model'] == bs[1]) & (test_df['mark'] == bs[0] ), 'model'] = 'other'
    
```

```python
test_df.groupby(['mark', 'model']).count()[test_df.groupby(['mark', 'model']).count()['id'] <= 5]['id']
```

```python
test_df.loc[test_df['model'] == 'other', 'modelEng'] = 'other'
test_df.loc[test_df['mark'] == 'other', 'markEng'] = 'other'
```

```python
test_df.to_csv('rare_cleaned_22.05.18.csv', encoding='utf-8', index=False)
```

## Check values that are not missing

```python
test_df.loc[:,['price', 'race', 'year','engineVolume']].describe()
```

```python
test_df.loc[(test_df['race'] > 1500), 'race'] = test_df.loc[(test_df['race'] > 1500), 'race'] // 1000
```

```python
test_df = test_df.drop(test_df[test_df['race'] > 1500].index)
```

```python
test_df = test_df.drop_duplicates()
```

```python
test_df.to_csv('rare_cleaned_22_05_18.csv', encoding='utf-8', index=False)
```
